Ejercicio 8

```python
# Ejercicio 8
uno = int(input("Ingresa un número entero: "))
dos = int(input("Ingresa otro número entero: "))

segundo = max(uno,dos)
primero = min(uno,dos)

suma = 0

for i in range(primero,segundo+1):
  suma += i
  if i != segundo:
    print(i, end=" + ")
  else:
    print(i, end=" ")

print("=",suma)
```
